```python
import numpy as np
def linear_threshold_gate(input_x: int,T:float) -> int:
  if input_x >= T:
    return 1
  else:
    return 0
```

```python
input_matrix = np.array ([
                          [0,0],
                          [0,1],
                          [1,0],
                          [1,1] ])
print(f'input table:\n {input_matrix}')
```

**AND GATE**


```python
weights = np.array([1,1])
print(f'weights:{weights}')
```

```python
#dot product of input and weights

dot_pr = np.dot(input_matrix,weights)
print(f'dot_product:{dot_pr}')
```

```python
T = 2
for i in range(0,4):
    activation = linear_threshold_gate(dot_pr[i],T)
    print(f'Activation: {activation}')
```

**OR GATE**

```python
weights = np.array([1,1])
print(f'weights:{weights}')
```

```python
dot_pr = np.dot(input_matrix,weights)
print(f'dot_product:{dot_pr}')
```

**NOR GATE**

```python
weights = np.array([-1,-1])
print(f'weights:{weights}')
```

```python
dot_pr = np.dot(input_matrix,weights)
print(f'dot_product:{dot_pr}')
```

```python
T = 0
for i in range(0,4):
    activation = linear_threshold_gate(dot_pr[i],T)
    print(f'Activation: {activation}')
```
